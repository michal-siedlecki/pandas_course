## S03 - L004 - Dodawanie i usuwanie kolumn

```python
import pandas as pd

fuel = pd.read_csv('course-files/fuel.csv', low_memory=False, \
                   usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'],\
                  index_col='Vehicle ID')
fuel.head()
```

```python
fuel['Combined KLP'] = fuel['Combined MPG (FT1)']* 1.609 / 3.78
fuel.head()
```

```python
fuel.insert(loc=4, column='liters per km', value=100/fuel['Combined KLP'])
```

```python
fuel.head()
```

```python
del fuel['Combined MPG (FT1)']
```

```python
fuel.head()
```

```python
fuel.drop(columns=['Combined KLP'], inplace=True)
```
