## S02 - L018 - Metoda map

```python
import pandas as pd
import random as rd

dictionary = {
    'PYT001' : 'Airbus 320',
    'PYT002' : 'Boeing 737',
    'PYT003' : 'Airbus 321'  
}

aircrafts = pd.Series(dictionary)

flightsList = []
for _ in range(100):
    flightsList.append(aircrafts.index[rd.randint(0,2)])
print(flightsList[:4])
```

```python
flights = pd.Series(flightsList)
```

```python
flights.head()
```

```python
flights_aircrafts = flights.map(aircrafts)
flights_aircrafts.head()
```
