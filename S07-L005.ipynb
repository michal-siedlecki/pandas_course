## S07-L005 - merge outer

```python
import numpy as np
import pandas as pd
```

```python
suppliers = pd.read_csv('course-files/northwind-mongo-master/suppliers _del_1.csv', \
                        usecols=['SupplierID','CompanyName', 'City'])
```

```python
products = pd.read_csv('course-files/northwind-mongo-master/products_del_1.csv',\
                      usecols=['ProductID','ProductName', 'SupplierID'])
```

```python
prod_and_sup = suppliers.merge(products, how='inner')
```

```python
len(suppliers)
```

```python
len(products)
```

```python
len(prod_and_sup)
```

```python
prod_and_sup_outer = suppliers.merge(products, how='outer', indicator=True)
```

```python
missed_left = prod_and_sup_outer['_merge'] == 'left_only'
```

```python
prod_and_sup_outer[missed_left]
```

```python
prod_and_sup_outer['_merge'].value_counts()
```

```python
missed_right = prod_and_sup_outer['_merge'] == 'right_only'
```

```python
prod_and_sup_outer[missed_right]
```
