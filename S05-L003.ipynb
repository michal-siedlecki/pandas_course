## S05-L003 - Transposition

```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('course-files/fuel.csv', usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'])
```

```python
fuel.set_index(['Year','Make','Model'], inplace=True)
```

```python
fuel.sort_index(inplace=True)
```

```python
fuel.head()
```

```python
fuel.loc[1984]
```

```python
fuel.loc[(1984,'Alfa Romeo')]
```

```python
fuel.loc[(1984,'Alfa Romeo','Spider Veloce 2000')]
```

```python
fuel.loc[(1984,'Alfa Romeo','Spider Veloce 2000'),'Range (FT1)']
```

```python
fuel = fuel.transpose()
```

```python
fuel.head()
```

```python
fuel.loc['Range (FT1)']
```

```python
fuel.loc['Range (FT1)',(1984,'Alfa Romeo','Spider Veloce 2000')]
```

```python
fuel = fuel.transpose()
```
