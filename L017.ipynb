```python
surveysTime = pd.read_csv('course-files/StackOverflowDeveloperSurvey2018.csv', usecols=['HoursOutside'], low_memory=False, squeeze=True)
surveysTime.head()
```

```python
surveysTime.value_counts()
```

```python
surveysTime = surveysTime.str.lower()
```

```python
surveysTime.head()
```

```python
surveysTime = surveysTime.apply(lambda old : str(old).upper())
surveysTime.head()
```

```python
def changeDescription(description):
    if description == 'LESS THAN 30 MINUTES':
        return 'LESS THAN HALF HOUR'
    return description

changeDescription('nan')
```

```python
changeDescription('LESS THAN 30 MINUTES')
```

```python
surveysTime.apply(changeDescription)
```
