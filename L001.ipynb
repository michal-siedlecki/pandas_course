```python
import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

days = np.array([1,2,3,4,5])
amounts = np.array([1,5,20,450,700])

plt.plot(days,amounts)
plt.xlabel('days')
plt.ylabel('amounts')
plt.show()
```

```python
plt.plot(days,amounts, 'bo')
plt.xlabel('days')
plt.ylabel('amounts')
plt.show()
```

```python
months = np.arange(1, 13, 1)
months
```

```python
income = np.array(100+3*months)
income
```

```python
plt.plot(months,income, 'go')
plt.xlabel('months')
plt.ylabel('income')
plt.show()
```

```python
cost = np.array(50+10*months)

plt.plot(months,income, 'go', months,cost, 'r^')
plt.xlabel('months')
plt.ylabel('income/cost')
plt.show()
```

```python
x = np.arange(-5., 5.1, 0.1)
x
```

```python
y = pow(2,x)

plt.plot(x,y)
plt.xlabel('Arguments')
plt.ylabel('Values')
plt.show()
```
