## S05-L005 - stack unstack

```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('course-files/fuel.csv', usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'])
```

```python
fuel.drop_duplicates(subset=['Year','Make','Model'], inplace=True)
```

```python
fuel.dropna(inplace=True)
```

```python
fuel.set_index(['Year','Make','Model'], inplace=True)
```

```python
fuel.head()
```

```python
fuel.stack()
```

```python
fuel.stack().to_frame()
```

```python
fuel.unstack('Model')
```

```python
fuel.unstack('Year')
```

```python
fuel.stack().unstack('Make')
```

```python
fuel.stack().unstack(['Year','Make'])
```

```python
fuel.stack().unstack(['Make','Year']).sort_index(axis=1)
```
