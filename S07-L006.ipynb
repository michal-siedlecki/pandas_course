## S07-L006 - left & right merge

```python
import numpy as np
import pandas as pd
```

```python
suppliers = pd.read_csv('course-files/northwind-mongo-master/suppliers.csv', \
                        usecols=['SupplierID','CompanyName', 'City'])
```

```python
products = pd.read_csv('course-files/northwind-mongo-master/products_del_2.csv',\
                      usecols=['ProductID','ProductName', 'SupplierID'])
```

```python
sup_prod = suppliers.merge(products, on='SupplierID', how='inner', indicator=True)
```

```python
len(sup_prod)
```

```python
len(products)
```

```python
len(suppliers)
```

```python
sup_prod_left = products.merge(suppliers, on='SupplierID', how='right', indicator=True)
```

```python
len(sup_prod_left)
```

```python
left_only = sup_prod_left['_merge'] == 'right_only'
```
