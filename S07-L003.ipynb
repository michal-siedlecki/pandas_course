## S07-L003 - metoda join

```python
import numpy as np
import pandas as pd
```

```python
customer_names = pd.read_csv('course-files/northwind-mongo-master/customers.csv',\
                             usecols=['CustomerID', 'CompanyName'])
```

```python
customer_cities = pd.read_csv('course-files/northwind-mongo-master/customers.csv',\
                             usecols=['CustomerID', 'City', 'Country'])
```

```python
len(customer_names)
```

```python
len(customer_cities)
```

```python
customer_names.set_index('CustomerID', inplace=True)
```

```python
customer_cities.set_index('CustomerID', inplace=True)
```

```python
customer_names.join(customer_cities)
```

```python
customer_names.reset_index(inplace=True)
```

```python
customer_names.join(customer_cities, on='CustomerID')
```
