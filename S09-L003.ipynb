## S09-L003 - excel

```python
import pandas as pd
import numpy as np
```

```python
mcdonalds = pd.read_csv('course-files/Mcdonalds.csv', usecols=['Category', 'Item', 'Calories', 'TotalFat'])
```

```python
ew = pd.ExcelWriter('mcdonalds-excel.xls')
```

```python
mcdonalds.to_excel(ew,index=False, sheet_name='Menu')
```

```python
ew.save()
```

```python
imported1 = pd.read_excel('mcdonalds-excel.xls', sheet_name='Menu')
```

```python
imported1.head()
```

```python
imported1['Calories'] = imported1['Calories']/2
```

```python
imported = imported1
```

```python
imported 
```

```python
imported.to_excel(ew, sheet_name='Menu_light', index=False)
```

```python
imported['Calories'] /= 2
```

```python
imported.to_excel(ew, sheet_name='Menu_extra_light', index=False)
```

```python
ew.save()
```
