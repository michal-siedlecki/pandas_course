```python
import numpy as np
import pandas as pd
```

```python
mrbean = pd.read_csv('course-files/mrbean_facebook_statuses_join.csv',\
                     usecols=['status_message_mrbean','status_type','num_likes'])
```

```python
nasa = pd.read_csv('course-files/nasa_facebook_statuses_join.csv',\
                   usecols=['status_message_nasa','status_type','num_likes'])
```
