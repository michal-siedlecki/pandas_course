## S08-L006 - Wykres scatter hexbin area

```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
marathon = pd.read_csv("course-files/marathon_results_2017.csv", usecols=["Age","M/F","Country","40K"])
marathon["TimeSeconds"] = marathon["40K"].apply(lambda x: pd.Timedelta(x).total_seconds())
marathon.head(5)
```

```python
marathon.plot.scatter(x='Age',y='TimeSeconds')
```

```python
marathon_m = marathon.loc[marathon['M/F']=='M']
```

```python
marathon_m.head()
```

```python
marathon_f = marathon.loc[marathon['M/F']=='F']
```

```python
marathon_f.head()
```

```python
ax = marathon_m.plot.scatter(x='Age', y='TimeSeconds', color='b', label='Men')
marathon_f.plot.scatter(x='Age', y='TimeSeconds', color='r', ax=ax, figsize=(15,10), label='Women')
```

```python
marathon_m.plot.hexbin(x='Age', y='TimeSeconds', label='Men', gridsize=10)
```

```python
marathon_m.plot.hexbin(x='Age', y='TimeSeconds', label='Men', gridsize=30)
```

```python
ax = marathon_f.plot.hexbin(x='Age', y='TimeSeconds', label='Men', gridsize=30)
marathon_m.plot.hexbin(x='Age', y='TimeSeconds', label='Men', gridsize=30, ax=ax)
```
