```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('course-files/fuel.csv', low_memory=False, index_col='Vehicle ID', \
                   usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'])
```

```python
fuel.head()
```

```python
fuel.info(memory_usage='deep')
```

```python
fuel['Year'] = fuel['Year'].astype('int32')
```

```python
fuel['Make'].value_counts()
```

```python
fuel['Make'] = fuel['Make'].astype('category')
```

```python
fuel.info(memory_usage='deep')
```

```python
cols = ['Model', 'Class', 'Fuel Type']

for elem in cols:
    print(fuel['{}'.format(elem)].value_counts())
    fuel['{}'.format(elem)] = fuel['{}'.format(elem)].astype('category')
```

```python
fuel.info(memory_usage='deep')
```

```python
fuel['Combined MPG (FT1)'] = fuel['Combined MPG (FT1)'].astype('float')
```

```python
fuel.info(memory_usage='deep')
```
