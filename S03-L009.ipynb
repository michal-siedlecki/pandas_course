## S03-L009-LAB Ranking

```python
import pandas as pd
import numpy as np

fortune = pd.read_csv('course-files/Fortune_500_2017.csv', index_col='Rank', low_memory=False,\
                   usecols=['Rank','Title','Employees','Profits','Assets'])
```

```python
fortune.head()
```

```python
fortune['Employees'].rank(ascending=False).head()
```

```python
fortune['RankByEmployee'] = fortune['Employees'].rank(ascending=False)
```

```python
fortune['RankByProfits'] = fortune['Profits'].rank(ascending=False)
```

```python
fortune.nlargest(3,'Assets')
```

```python
fortune.nsmallest(3,'Assets')
```

```python
fortune.nlargest(20,['RankByEmployee', 'RankByProfits'])
```
