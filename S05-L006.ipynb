## S05-L006 - pivot_table

```python
import pandas as pd
import numpy as np
```

```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('course-files/marathon_results_2016.csv', usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],)
```

```python
df.set_index('Bib', inplace=True)
```

```python
df.head()
```

```python
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
```

```python
df.info()
```

```python
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
```

```python
df.head()
```

```python
df.info()
```

```python
df.pivot_table(values='TotalSeconds',columns='M/F', index='Age')
```

```python
df.pivot_table(values=['TotalSeconds','HalfSeconds'],columns='M/F', index='Age')
```

```python
dfPivot = df.pivot_table(values=['TotalSeconds','HalfSeconds'],columns='M/F', index='Age')
```

```python
dfPivot = dfPivot.swaplevel(axis=1).sort_index(axis=1)
```

```python
dfPivot.head()
```

```python
df.head()
```

```python
df.pivot_table(values='City', index='Age', columns='M/F', aggfunc='count')
```

```python
df.pivot_table(values='City', index='Age', aggfunc='count')
```

```python
df.pivot_table(values='City', index='Country', aggfunc='count').sort_values(['City'], ascending=False)
```
