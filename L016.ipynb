## S02 - L016 - Więcej funkcji Data Series 

```python
import pandas as pd

programmers = pd.read_csv('course-files/StackOverflowDeveloperSurvey2018.csv', low_memory=False, usecols=['ConvertedSalary'], squeeze=True)
programmers.dropna(inplace=True)
programmers.head(10)
```

```python
programmers.mean()
```

```python
programmers.median()
```

```python
programmers.std()
```

Odchylenie standartowe czterokrotnie przewyższa medianę - zarobki są bardzo zróżnicowane. Jest wielu programistów zarabiających przeciętnie ale niewielka grupa zarabiająca bardzo dużo zawyża średnią zarobków.

```python
programmers.max()
```

```python
programmers.min()
```
