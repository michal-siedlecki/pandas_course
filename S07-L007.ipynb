## S07-L007 - opcje polecenia merge

```python
import numpy as np
import pandas as pd
```

```python
suppliers = pd.read_csv('course-files/northwind-mongo-master/suppliers.csv', \
                        usecols=['SupplierID','CompanyName', 'City'])
```

```python
products = pd.read_csv('course-files/northwind-mongo-master/products.csv',\
                      usecols=['ProductID','ProductName', 'SupplierID'])
```

```python
suppliers.set_index('SupplierID', inplace=True)
```

```python
suppliers.merge(products, left_index=True, right_on='SupplierID').head()
```

```python
products.set_index('SupplierID', inplace=True)
```

```python
suppliers.merge(products, left_index=True, right_index=True).head()
```

```python
suppliers.reset_index(inplace=True)
```

```python
suppliers.rename({'SupplierID': 'supplierId'},axis=1, inplace=True)
```
